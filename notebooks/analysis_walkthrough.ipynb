# Population Growth Analysis Pipeline

This notebook walks through the full analysis pipeline: loading data, preprocessing, visualization, and forecasting.

```python
import sys
import os
sys.path.append('..')

from src.data_loader import WorldBankLoader
from src.preprocess import preprocess_population_data
from src.visualize import *
from src.analysis import PopulationAnalyzer
import plotly.io as pio
pio.renderers.default = 'notebook'
```

## 1. Load Data
Fetching data for G20 nations.

```python
g20_codes = ["ARG", "AUS", "BRA", "CAN", "CHN", "FRA", "DEU", "IND", "IDN", "ITA", 
             "JPN", "MEX", "RUS", "SAU", "ZAF", "KOR", "TUR", "GBR", "USA", "EUU"]
loader = WorldBankLoader(cache_dir="../data")
raw_df = loader.fetch_multiple_countries(g20_codes)
df = preprocess_population_data(raw_df)
df.head()
```

## 2. Visualize Trends
Plotting the population of top G20 countries.

```python
plot_population_trends(df, ["India", "China", "United States", "Indonesia"])
```

## 3. Growth Comparison
Heatmap of growth rates by decade.

```python
plot_growth_heatmap(df)
```

## 4. Forecasting India vs China to 2075

```python
analyzer = PopulationAnalyzer(df)
india_pred = analyzer.predict_future("India", horizon_years=55)
china_pred = analyzer.predict_future("China", horizon_years=55)

import matplotlib.pyplot as plt
plt.plot(df[df['country'] == 'India']['year'], df[df['country'] == 'India']['population'], label='India Actual')
plt.plot(india_pred['year'], india_pred['log_prediction'], '--', label='India Forecast (Logistic)')
plt.plot(df[df['country'] == 'China']['year'], df[df['country'] == 'China']['population'], label='China Actual')
plt.plot(china_pred['year'], china_pred['log_prediction'], '--', label='China Forecast (Logistic)')
plt.legend()
plt.title("India vs China Population Forecast")
plt.show()
```
